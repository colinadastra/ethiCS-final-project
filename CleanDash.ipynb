```python
#minimal example for debugging
from IPython import get_ipython
from IPython.display import Markdown, display
import nbformat

# Load Dashboard Notebook
with open("CleanDashTester.ipynb", "r") as f:
#with open("Dashboard.ipynb", "r") as f:
    dashboard_nb = nbformat.read(f, as_version=4)
    

# Function to display a specific markdown cell by index
# Function to display a specific markdown cell by index
def display_markdown_cell(index):
    cell = dashboard_nb.cells[index]
    if cell.cell_type == "markdown":
        try:
            # Destroy the old Markdown import
            #del Markdown
            # Re-import Markdown to ensure it's in the correct state
            from IPython.display import Markdown
            display(Markdown(cell.source))
        except Exception as e:
            print(f"Fudge! Error displaying markdown cell {index}: {e}")

# Function to execute a specific code cell by index
def execute_code_cell(index):
    cell = dashboard_nb.cells[index]
    if cell.cell_type == "code":
        try:
            get_ipython().run_cell(cell.source)
        except Exception as e:
            print(f"Error executing code cell {index}: {e}")

# Example usage:
# Display markdown cells
##display_markdown_cell(0)
display_markdown_cell(2)

# Execute code cells
#execute_code_cell(1)
#execute_code_cell(3)
# Why do Markdown cells stop working after the code executes! Is it the interactive widgets?
#display_markdown_cell(2)

# Continue with other cells as needed

#CleanDashTester.ipynb - a loop to call all cells in the notebook
execute_code_cell(0)
display_markdown_cell(1)
execute_code_cell(2)
display_markdown_cell(3)
```
